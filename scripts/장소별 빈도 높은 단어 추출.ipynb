## 3.2 데이터 단어 추출 함수

```python
import re
from konlpy.tag import Okt
from collections import Counter

def text_cleaning(text):
    hangle = re.compile('[^ ㄱ-ㅣ 가-힣]') # 정규표현식 제거 단어
    result = hangle.sub('', text)        # 정규표현식 제거 단어 적용
    okt = Okt()                            # 단어 형태소 분석 모델
    nouns = okt.nouns(result)              # 단어 형태소 분석
    nouns = [x for x in nouns if x not in  stopwords]      # 불용어 제거
    return nouns
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def word_filter(data):
    vect = CountVectorizer(tokenizer = lambda x: text_cleaning(x))
    bow_vect = vect.fit_transform(data["리뷰"].tolist())
    word_list = vect.get_feature_names()
    count_list = bow_vect.toarray().sum(axis =0)
    word_info = pd.concat([pd.DataFrame(word_list),pd.DataFrame(count_list)],axis = 1)
    word_info.columns = ["단어","빈도수"]
    word_info.sort_values("빈도수", ascending = True)
    return word_info
```
